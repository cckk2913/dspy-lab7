```python
# 用 dict 創造 DF
## 字典的 key 會變成欄位名稱，value 則是欄位的值

data = {
  'school': ["台大", "清大", "陽明交大","成大"],
  'year': [1928, 1925, 2021, 1931],
  'students': [30000, 20000, 10000, 40000]
}

df = pd.DataFrame(data)
df
```

```python
# 用pd.read_csv()讀取檔案

import pandas as pd

df = pd.read_csv('./movie.csv', encoding='utf-8')
```

```python
df.shape
```

```python
df.head()
```

<mark> df.iloc[] ➡️ Select row by row index </mark>

```python
df.iloc[1]     # Select Row by row index
```

```python
df.iloc[:3]    # Select First 3 Rows
```

```python
df.iloc[-1:]   # Select Last Row
```

```python
df.iloc[-3:]   # Select Last 3 Row
```

<mark> df.loc[] ➡️ Select column by 欄位名 / 欄位index </mark>

```python
# Using loc[] to take column slices
df2 = df.loc[:, ["Courses","Fee","Duration"]] # Selecte multiple columns
df2 = df.loc[:, 'Fee':'Discount']             # Select columns between two columns

# Using iloc[] to select column by column index
df2 = df.iloc[:,[1,3,4]]  # Select columns by column index
df2 = df.iloc[:,1:4]      # Select between indexes 1 and 4 (2,3,4)
df2 = df.iloc[:,2:]       # Select From 3rd column to the last column
df2 = df.iloc[:,:2]       # Select First 2 columns
```
