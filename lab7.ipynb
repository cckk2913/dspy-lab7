<mark>篩出[符合特定條件]的資料</mark>

`df.query(表達式, inplace=False, **kwargs)`

- `.query(表達式)` 會依據表達式寫的條件，**篩選出DataFrame中 符合條件的 row 。**

- 這個表達式可以包含一個或多個條件。

  > **表達式** : Expression which takes conditions to query rows
  
  > **inplace** : Defaults to `False`. When set to  `True`, it updates the referring DataFrame and query() method returns None.
  
  > **\*\*kwargs** :  Keyword arguments that works with eval()
  



````python
# 格式：df.query('欄位名 == 值')
## 篩選出 name_TW欄 是 "鋼鐵人" 的 rows 

df.query('name_TW == "鋼鐵人"')    
````

## 另一系列篩選寫法

**`df[ df[欄位名] == 值 ]`**

例如：df[ df['name_TW'] == '鋼鐵人' ]

> 有兩個中括號 [ ]

> 裡面的 `df['name_TW'] == '鋼鐵人'`產生一個布林值的 Series，用這個 Series 去篩選 df

像是：

| name_TW | `df['name_TW'] == '鋼鐵人'` 產出的篩網 | 最終回傳的結果 |
| -------- | -------- | -------- |
| 鋼鐵人     | True     |  ➡️ 會出現 |
| 鋼鐵人     | True     |   ➡️ 會出現 |
| 你的名字   | False     |      x             |
| 鋼鐵人     | True     |   ➡️ 會出現 |

````python
df[df["Date"] == '2015-05-19'] 
````

````python
# 剛剛做過的事情，用 .loc 也可以做到

values = ['Iron Man', '你的名字']

df.loc[df['name_EN'] == 'Your Name']
df.loc[df['name_EN'] != 'Your Name']
df.loc[df['name_EN'].isin(values)]
df.loc[~df['name_EN'].isin(values)]
df.loc[(df['Star'] >= 3) & (df['Star'] <= 5)]
df.loc[(df['Star'] >= 3) & (df['Date'] == '2015-05-19' )]

# 根據字串的條件篩選
df[df["name_EN"] == 'The Great Wall'] 
df[df['name_EN'].str.contains("Great")]
df[df['name_EN'].str.lower().str.contains("avenger")]
df[df['name_EN'].str.startswith("A")]
````

## Iterate over a DataFrame

### 1. 遍歷 Rows

使用 `iterrows()` ，會回傳一個 tuple，第一個值是 row index，第二個值是 row value 的 Series，

所以下面我們用 for loop 時，可以用兩個變數 `id, row` 去接它們：

````python
df = pd.read_csv('./movie.csv', encoding='utf-8')
df = df[:5] # 為了版面簡潔，我們先取前5筆資料就好

# 遍歷每一列 
for id, row in df.iterrows():
  print('這一列index是：', id)
  print(row)
  print('🐳🐳🐳🐳🐳')
````

上面的例子for 迴圈遍歷 DF，變數 row 包含了一整列的資料

<mark>但我只想針對特定欄位欸？</mark>


````python
df = pd.read_csv('./movie.csv', encoding='utf-8')
df = df[:5] 
````

````python
# 遍歷每一列，但只取出特定欄位的值
for id, row in df.iterrows():
  print('這列的 name_TW 值：', row['name_TW'])
  print('這列的 Star 值：', row['Star'])
  print('🐳🐳🐳🐳🐳')
````

````python
# 做較複雜的事情（if[星星]大於3 就印出[該電影]好看，else印出[該電影]不推薦）

for i, row in df.iterrows():
  if row['Star'] > 3:
    print(row['name_TW'], '🌟🌟🌟好看')
  else:
    print(row['name_TW'], '🆘不推薦')
````

### 3. 將函數應用到columns 或rows

使用 .apply() 將某個函數應用在 DataFrame 的每一行或每一列。  

````python
# 先定義一個函數：將傳入的值乘以2
def double_value(x):
    return x * 2
````

## df.groupby()

將 DataFrame 的rows分組，並對每一組進行操作。此函數在資料分析很重要


<mark>.groupby()的參數包含：</mark>

```python
df.groupby(by=None, axis=0, level=None, as_index=True, 
       sort=True, group_keys=True, squeeze=<no_default>, 
       observed=False, dropna=True)

    by – 要依據哪一column底下的資料進行分組
    axis – Default to 0. It takes 0 or ‘index’, 1 or ‘columns’
    level – Used with MultiIndex.
    as_index – sql style grouped otput.
    sort – Default to True. 指定是否在分組後要排序資料
    group_keys – add group keys or not
    squeeze – depricated in new versions
    observed – This only applies if any of the groupers are Categoricals.
    dropna – Default to False. Use True to drop None/Nan on sory keys
```

````python
from pprint import pprint #列印比較好看
import pandas as pd

# 先讀取資料（前20筆）
df = pd.read_csv('./movie.csv', encoding='utf-8')
df = df[:20] 
````

````python
# 根據'Star'欄位，將 rows 分組
grouped = df.groupby(by='Star')

# 這是 DataFrameGroupBy object
grouped
````

````python
# 用 .groups 才能看到這個DataFrameGroupBy object分了哪幾組、每組包含哪些row
pprint(grouped.groups)
````

````python
'''
上面是這樣把 20 筆資料分組的：
1顆星的rows 包含： [4, 19]（index是4和9的rows，也就是第5和第20個橫列）
2顆星的rows 包含： [11, 15]
3顆星的rows 包含： [1, 3, 6, 7, 10, 12, 13]
4顆星的rows 包含： [8, 16]
5顆星的rows 包含： [0, 2, 5, 9, 14, 17, 18]
'''
````

````python
# 用 .size() 看每組分別有幾個rows
grouped.size()
````

````python
# 用 .get_group() 取出特定組的rows
## 例如取出 4 顆星 的rows
grouped.get_group(4)
````

````python
# 用多欄位分組
df.groupby(['Star', 'name_EN']).groups
````

<mark> 用 groupby 做一些統計</mark>

````python
# 用'電影名字'分組，並取得每組的加總（會忽略非數字的欄位）
grouped = df.groupby(by='name_EN')
grouped.sum()
````

但每個電影的評論數不同，把星星數加總不合理，所以要算的是<mark>星星數平均值</mark>

````python
# 用'電影名'分組，看看每組（每個電影）的'Star' 平均數
## 上面例子的 .sum() 換成 .mean()

grouped = df.groupby(by='name_EN')
average_rating = grouped['Star'].mean()
average_rating
````

````python
# 按照平均星星分數，由高到低排序

average_rating.sort_values(ascending=False)
````

````python
# 畫出來
average_rating.plot(kind='bar')
````
