## Iterate over a DataFrame

### 1. éæ­· Rows

ä½¿ç”¨ `iterrows()` ï¼Œæœƒå›å‚³ä¸€å€‹ tupleï¼Œç¬¬ä¸€å€‹å€¼æ˜¯ row indexï¼Œç¬¬äºŒå€‹å€¼æ˜¯ row value çš„ Seriesï¼Œ

æ‰€ä»¥ä¸‹é¢æˆ‘å€‘ç”¨ for loop æ™‚ï¼Œå¯ä»¥ç”¨å…©å€‹è®Šæ•¸ `id, row` å»æ¥å®ƒå€‘ï¼š

````python
# ç‚ºäº†ç‰ˆé¢ç°¡æ½”ï¼Œæˆ‘å€‘å…ˆå–å‰5ç­†è³‡æ–™
df = pd.read_csv('./movie.csv', encoding='utf-8')
df = df[:5] 
````

````python
# éæ­·æ¯ä¸€åˆ— 
for id, row in df.iterrows():
  print('é€™ä¸€åˆ—indexæ˜¯ï¼š', id)
  print(row)
  print('ğŸ³ğŸ³ğŸ³ğŸ³ğŸ³')
````

ä¸Šé¢çš„ä¾‹å­for è¿´åœˆéæ­· DFï¼Œè®Šæ•¸ row åŒ…å«äº†ä¸€æ•´åˆ—çš„è³‡æ–™

<mark>ä½†æˆ‘åªæƒ³é‡å°ç‰¹å®šæ¬„ä½æ¬¸ï¼Ÿ</mark>


````python
# éæ­·æ¯ä¸€åˆ—ï¼Œä½†åªå–å‡ºç‰¹å®šæ¬„ä½çš„å€¼
for id, row in df.iterrows():
  print('é€™åˆ—çš„ name_TW å€¼ï¼š', row['name_TW'])
  print('é€™åˆ—çš„ Star å€¼ï¼š', row['Star'])
  print('ğŸ³ğŸ³ğŸ³ğŸ³ğŸ³')
````

````python
# åšè¼ƒè¤‡é›œçš„äº‹æƒ…
# if[æ˜Ÿæ˜Ÿ]å¤§æ–¼3å°±å°å‡º[è©²é›»å½±]å¥½çœ‹ï¼Œelseå°å‡º[è©²é›»å½±]ä¸æ¨è–¦

for i, row in df.iterrows():
  if row['Star'] > 3:
    print(row['name_TW'], 'ğŸŒŸğŸŒŸğŸŒŸå¥½çœ‹')
  else:
    print(row['name_TW'], 'ğŸ†˜ä¸æ¨è–¦')
````

````python
# å…ˆå®šç¾©ä¸€å€‹å‡½æ•¸ï¼šå°‡å‚³å…¥çš„å€¼ä¹˜ä»¥2
def double_value(x):
    return x * 2
````
