  æˆ‘å€‘æœ‰åˆ—è¡¨  `mylist = ['aaaa', 'rrr', '123']`ï¼Œè«‹ä¾åºæŠŠåˆ—è¡¨ä¸­çš„å­—ä¸²åˆ°å‡½æ•¸ã€‚

  å‡½æ•¸è¦æ±‚ï¼š

   - å…¶ä»–æƒ…æ³ï¼Œå‡½æ•¸è¦å›å‚³ 'éƒ½æ²’æœ‰'

æŠŠ 3 æ¬¡å›å‚³çš„çµæœå­˜åˆ°åˆ—è¡¨ `result_list` ä¸­ï¼Œä¸¦å°å‡º `result_list`ã€‚

```python
import re
'''é€™å€‹å¥—ä»¶ï¼ˆregular expressionï¼‰å¯ä»¥å¹«æˆ‘å€‘è™•ç†æ–‡å­—
re.search() å¯ä»¥æœå°‹å­—ä¸²è£¡é¢ï¼Œæœ‰æ²’æœ‰æˆ‘å€‘è¦æ‰¾çš„æŸå€‹æ±è¥¿'''


## åœ¨é€™è£¡å®šç¾©å‡½æ•¸
def check(text):
  if re.search(r'[abc]', text):
    return 'åŒ…å«abc'
  elif re.search(r'[0-9]', text):
    return 'åŒ…å«æ•¸å­—'
  else:
    return 'æ²’æœ‰abcï¼Œä¹Ÿæ²’æœ‰æ•¸å­—'
    
    


## ç”¨è¿´åœˆéæ­· mylistï¼Œä¾åºæŠŠä¸‰å€‹å­—ä¸²å‚³çµ¦ check() 
## æ¯æ¬¡çš„å›å‚³çµæœè¦åŠ å…¥ result_list 


mylist = ['aaaa', 'rrr', '123']

result_list = []

for m in mylist:
  result = check(m)
  result_list.append(result)
  

print(result_list)
```

```python
data = {
  'school': ["å°å¤§", "æ¸…å¤§", "é™½æ˜äº¤å¤§","æˆå¤§"],
  'year': [1928, 1925, 2021, 1931]
}

df = pd.DataFrame(data)
df
```

```python
# loc
df.loc[len(df)] = ['æ–°æ ¡', 2023]
df
```

```python
# concat: çµåˆå…©å€‹ df ï¼ˆæˆ‘å€‘æŠŠæ–°çš„row ç›´æ¥è®Šæˆä¸€å€‹åªæœ‰ä¸€åˆ—è³‡æ–™çš„df)

new_row = pd.DataFrame({'school': ['NTU'], 'year': [1928]})
df = pd.concat([df, new_row], ignore_index=True)
```

```python
# _append
''' 
In the new version of Pandas, the `append` method is changed to `_append`
The `ignore_index=True` argument ensures that the index of the newly added row is reset to maintain a continuous integer index.
'''

new_row = {'school': 'æ–°æ–°æ ¡', 'year': 2023}
df = df._append(new_row, ignore_index=True)
df
```

### <mark>DF è³‡æ–™æ¸…ç†ï¼šç©ºå€¼</mark>

- ç©ºå€¼æœƒå½±éŸ¿æˆ‘å€‘çš„åˆ†æçµæœï¼Œæ‰€ä»¥è¦å…ˆè™•ç†ç©ºå€¼

- ä¸€ç¨®æ–¹æ³•æ˜¯ç›´æ¥åˆªé™¤`åŒ…å«ç©ºå€¼çš„row`ï¼Œå› ç‚ºè³‡æ–™å¯èƒ½éå¸¸å¤§ï¼Œåˆªé™¤ä¸€äº›åˆ—ä¸æœƒç”¢ç”Ÿå¾ˆå¤§å½±éŸ¿ã€‚

```python
import pandas as pd
data = pd.read_csv('./iris.csv')

# åŸå§‹è³‡æ–™çš„é•·åº¦ï¼ˆrowæ•¸é‡ï¼‰
len(data)
```

```python
# ç”¨ .dropna() åˆªé™¤æœ‰ç©ºå€¼çš„ row
## .dropna() æœƒè¿”å›ä¸€å€‹æ–°çš„DFï¼Œä¸æœƒæ›´æ”¹åŸä¾†çš„DF
## æ‰€ä»¥è¦å®šç¾©ä¸€å€‹æ–°çš„è®Šæ•¸(new_df)ä¾†å­˜æ”¾

new_df = data.dropna()
```

```python
# ç¢ºèªrowæ•¸é‡æ˜¯å¦æ¸›å°‘
len(new_df)
```

```python
# å¦‚æœè¦æ›´æ”¹åŸä¾†çš„DFï¼Œå¯ä»¥åŠ ä¸Šåƒæ•¸ inplace = True
## é€™æ¨£å°±ä¸ç”¨å®šç¾©æ–°çš„è®Šæ•¸ä¾†å­˜æ”¾
## ä½†å°±ç„¡æ³•å›å¾©åŸä¾†çš„DFäº†ï¼ˆé™¤éé‡æ–°è®€å–æª”æ¡ˆï¼‰
data.dropna(inplace = True)
len(data)
```

- å¦ä¸€ç¨®è™•ç†ç©ºå€¼çš„æ–¹æ³•æ˜¯`å¡«å…¥æ–°çš„å€¼`ï¼Œä¾‹å¦‚å¡«å…¥0æˆ–å¹³å‡å€¼ï¼Œé€™æ¨£å°±ä¸å¿…åˆªé™¤æ•´å€‹row

```python
import pandas as pd

data = pd.read_csv('./iris.csv')

data.fillna(0, inplace = True) 
len(data)
```

```python
# æœ¬ä¾†æœ‰ç©ºå€¼çš„æ¬„ä½ï¼Œç¾åœ¨éƒ½è¢«å¡«ä¸Š 0 äº†ï¼Œä¹Ÿå°±éƒ½æœ‰ 150 å€‹å€¼äº†
data.info()
```

```python
# ä¸Šé¢çš„ç¯„ä¾‹æ›¿æ›äº†æ•´å€‹ data ä¸­çš„æ‰€æœ‰ç©ºç™½å„²å­˜æ ¼
# åªæƒ³æ›¿æ›æŸä¸€å€‹ç‰¹å®š column åº•ä¸‹çš„ç©ºç™½å€¼ï¼š
data['SepalLengthCm'].fillna(0, inplace = True)
```

- æ›¿æ›ç©ºç™½å„²å­˜æ ¼çš„å¸¸è¦‹æ–¹æ³•ï¼Œæ˜¯å¡«å…¥ column çš„å¹³å‡å€¼ã€ä¸­ä½æ•¸æˆ–çœ¾æ•¸ï¼Œè€Œä¸æ˜¯0
- ä½¿ç”¨ mean()ã€median() å’Œ mode() ä¾†è¨ˆç®—

```python
import pandas as pd
df = pd.read_csv('./iris.csv')

# è¨ˆç®— SepalLengthCm æ¬„çš„å¹³å‡å€¼ï¼Œä¸¦å­˜åˆ°è®Šæ•¸ x
x = df["SepalLengthCm"].max()

# ç”¨ x æ›¿æ› SepalLengthCm æ¬„çš„ç©ºç™½å€¼
df["SepalLengthCm"].fillna(x, inplace = True) 
```
