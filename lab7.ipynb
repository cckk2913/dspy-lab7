````python
df = pd.read_csv('./movie.csv', encoding='utf-8')
df = df[:5] 
````

````python
# éæ­·æ¯ä¸€åˆ—ï¼Œä½†åªå–å‡ºç‰¹å®šæ¬„ä½çš„å€¼
for id, row in df.iterrows():
  print('é€™åˆ—çš„ name_TW å€¼ï¼š', row['name_TW'])
  print('é€™åˆ—çš„ Star å€¼ï¼š', row['Star'])
  print('ğŸ³ğŸ³ğŸ³ğŸ³ğŸ³')
````

````python
# åšè¼ƒè¤‡é›œçš„äº‹æƒ…ï¼ˆif[æ˜Ÿæ˜Ÿ]å¤§æ–¼3 å°±å°å‡º[è©²é›»å½±]å¥½çœ‹ï¼Œelseå°å‡º[è©²é›»å½±]ä¸æ¨è–¦ï¼‰

for i, row in df.iterrows():
  if row['Star'] > 3:
    print(row['name_TW'], 'ğŸŒŸğŸŒŸğŸŒŸå¥½çœ‹')
  else:
    print(row['name_TW'], 'ğŸ†˜ä¸æ¨è–¦')
````

